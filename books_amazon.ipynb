```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/OscarL7/Amazon-bestsellers-books/main/bestsellers%20with%20categories.csv')
df.head()
```

```python
# 1. Distribution of User Ratings
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='User Rating', kde=True)
plt.title('Distribution of User Ratings')
plt.xlabel('User Rating')
plt.ylabel('Count')
plt.show()
```

```python
# 2. Top Authors by Number of Bestsellers
plt.figure(figsize=(10, 6))
top_authors = df['Author'].value_counts().nlargest(10)
sns.barplot(x=top_authors, y=top_authors.index, orient='h')
plt.title('Top Authors by Number of Bestsellers')
plt.xlabel('Number of Bestsellers')
plt.ylabel('Author')
plt.show()
```

```python
# 3. Genre Comparison
genre_counts = df['Genre'].value_counts()
plt.figure(figsize=(10, 6))
sns.barplot(x=genre_counts.index, y=genre_counts.values)
plt.title('Genre Comparison')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.show()
```

```python
# 4. Price Analysis
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Price', y='User Rating')
plt.title('Price Analysis')
plt.xlabel('Price')
plt.ylabel('User Rating')
plt.show()
```

```python
# 5. Reviews vs. Ratings
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Reviews', y='User Rating')
plt.title('Reviews vs. Ratings')
plt.xlabel('Reviews')
plt.ylabel('User Rating')
plt.show()
```

```python
# 6. Price Distribution
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, y='Price')
plt.title('Price Distribution')
plt.ylabel('Price')
plt.show()
```
